# **DECISION TREES**

- Supervised Algorithm used to Classification or Regression;
- Trees Representative. The trees have 3 knots like root, branch and sheets;
- Roof and Branch are ours check points;
- The algorithm make a decision on each knots;
- The sheets are our final result.
DDT.PNG



- It's support all kind of distinct data (Categorical and Numeric);
- You don't need to use Pre-Processing of data like distance models;
- Better against outliers;
- Willing to overfitting: We need to reduce its growth, limiting the maximum depth, its maximum features value, its maximum sheet levels, verifying its information earnings from an attributes during tree builds and cut it or cut branchs after build our tree.
- Recommend to weak relation between dependent and independent variables;
- Necessary for understanding and generating information;
- Which are our attributes that are used like roof and branchs?
- How choose them? Using purity calculation, more uniform it is, better will be.
- When does tree have to stop?

### **Navegating through all decision tree structure**

```
from sklearn.tree import export_graphviz
from sklearn.datasets import load_iris
from sklearn import tree
import numpy as np
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
```

```
iris = load_iris()
X = iris.data
y = iris.target
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0)
```

```
estimator = DecisionTreeClassifier(max_leaf_nodes=3, random_state=0)
estimator.fit(X_train, y_train) # training our object
```

```
# Atributs that we want to see
n_nodes = estimator.tree_.node_count
children_left = estimator.tree_.children_left
children_right = estimator.tree_.children_right
feature = estimator.tree_.feature
threshold = estimator.tree_.threshold
```

```
print("Número de nós: {}".format(n_nodes))
print("Indices a esquerda: {}".format(children_left))
print("Indices a direita: {}".format(children_right))
print("Feature: {}".format(feature))
print("Limite (threshold: {}".format(threshold))
```

```
node_depth = np.zeros(shape=n_nodes, dtype=np.int64)
is_leaves = np.zeros(shape=n_nodes, dtype=bool)
stack = [(0, -1)]
while len(stack) > 0:
    node_id, parent_depth = stack.pop()
    node_depth[node_id] = parent_depth + 1

    if (children_left[node_id] != children_right[node_id]):
        stack.append((children_left[node_id], parent_depth + 1))
        stack.append((children_right[node_id], parent_depth + 1))
    else:
        is_leaves[node_id] = True

print("\nA arvore binária tem %s nós e a seguinte estrutura: \n"
      % n_nodes)
for i in range(n_nodes):
    if is_leaves[i]:
        print("%snó=%s (nó folha)." % (node_depth[i] * "\t", i))
    else:
        print("%snó=%s (nó teste): vai para o nó %s se o valor do atributo %s <= %s \n se não, vai para o "
              "nó %s."
              % (node_depth[i] * "\t",
                 i,
                 children_left[i],
                 iris.feature_names[feature[i]],
                 threshold[i],
                 children_right[i],
                 ))
```

**Extracting rules from tree generated from X_test**

```
def extrai_regras(sample_id):
    node_indicator = estimator.decision_path(X_test)

    leave_id = estimator.apply(X_test)

    #sample_id = sample
    node_index = node_indicator.indices[node_indicator.indptr[sample_id]:
                                        node_indicator.indptr[sample_id + 1]]

    print('\nRegras usadas para predizer a amostra %s' % (sample_id))

    for f,v in zip(iris.feature_names,X_test[sample_id]):
        print('%s = %s'%(f,v))

    print('\n')      
    for node_id in node_index:
        if leave_id[sample_id] == node_id:
            continue

        if (X_test[sample_id, feature[node_id]] <= threshold[node_id]):
            threshold_sign = "<="
        else:
            threshold_sign = ">"

        print("id do nó de decisão %s : (atributo %s com valor = %s %s %s)"
              % (node_id,
                 iris.feature_names[node_id],
                 X_test[sample_id, feature[node_id]],
                 threshold_sign,
                 threshold[node_id]))
    pred =estimator.predict(X_test[sample_id].reshape(1, -1)) 
    print("\tClasse => %s" %iris.target_names[pred])
```

```
extrai_regras(1)
```

```
extrai_regras(2)
```

*This parameter controls the attribute used to divide the tree. So, as we reduce the min_samples, the tree will grow. With min_sample_split we must have  at least 1 knot with 2 branchs. Like we see the red circule is where we cut the tree.* 

```
from mlxtend.plotting import plot_decision_regions
import matplotlib.pyplot as plt
```

```
def visualize_fronteiras(msamples_split):
    X = iris.data[:, [0, 2]]
    y = iris.target

    clf = DecisionTreeClassifier(min_samples_split=msamples_split)
    tree = clf.fit(X, y)

    plt.figure(figsize=(8,5))
    plot_decision_regions(X, y, clf=tree, legend=2)

    plt.xlabel('sepal length [cm]')
    plt.ylabel('petal length [cm]')
    plt.title('Decision Tree on Iris')
    plt.show()
```

```
# At minimum 2 samples below
visualize_fronteiras(2)
```
